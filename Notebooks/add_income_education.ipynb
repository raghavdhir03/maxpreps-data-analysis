```python
df = pd.read_csv('zcta_data.csv')
```

### Extract Zipcode and calculate the percent of 25+ year olds with Bachelor's Degree

```python
# Extract ZIP codes from the 'name_e' column by splitting the string and keeping the second part
df['ZIP Code'] = df['name_e'].str.split().str[1]

# Filter the relevant columns: extracted ZIP code, total population 25+, and bachelor's degree population
df = df[['ZIP Code', 'aqpke001', 'aqpke022', 'aqq4e001']]

# Rename columns for better understanding
df.columns = ['ZIP Code', 'Total Population 25+', 'Bachelor\'s Degree Population', 'Median Household Income']

# Calculate the percentage of 25+ population with a bachelor's degree
df['Percent with Bachelor\'s Degree'] = (df['Bachelor\'s Degree Population'] / df['Total Population 25+']) * 100

df
```

### Merge pct Bachelor data onto final datset

```python
#right = df.rename(columns={"Percent with Bachelor's Degree": 'Team 1 Bachelor Degree', 'Median Household Income': 'Team 1 Median Household Income'}).drop(columns=['Total Population 25+', 'Bachelor\'s Degree Population'])
right = df.rename(columns={'Median Household Income': 'Team 1 Median Household Income'}).drop(columns=['Total Population 25+', 'Bachelor\'s Degree Population', 'Percent with Bachelor\'s Degree'])
right['ZIP Code'] = right['ZIP Code'].astype(int)
```

```python
#perfrom the merge to get team 1 bachelors data
final = pd.merge(final, right, left_on='Team 1 Zipcode', right_on='ZIP Code', how = 'left')
final
```

```python
#get team 2 bachelors data
#right = df.rename(columns={"Percent with Bachelor's Degree": 'Team 2 Bachelor Degree'}).drop(columns=['Total Population 25+', 'Bachelor\'s Degree Population'])
right = df.rename(columns={'Median Household Income': 'Team 2 Median Household Income'}).drop(columns=['Total Population 25+', 'Bachelor\'s Degree Population', 'Percent with Bachelor\'s Degree'])
right['ZIP Code'] = right['ZIP Code'].astype(int)
```

```python
#perfrom the merge to get team 2 bachelors data
final = pd.merge(final, right, left_on='Team 2 Zipcode', right_on='ZIP Code', how = 'left')
final
```
