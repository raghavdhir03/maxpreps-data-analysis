```python
# clean up the names for each column
cols = ['School Name', 'State',
       'Address',
       'City',
       'Zipcode',
       'State Abbr',
        'School Type',
        'Total Students All Grades',
        'Free and Reduced Lunch',
        'Grades 9-12 Students',
       'American Indian/Alaska Native',
       'Asian or Asian/Pacific Islander',
       'Hispanic',
       'Black',
       'White',
       'Nat. Hawaiian or Other Pacific Isl.',
       'Two or More Races',
       'Pupil/Teacher Ratio',
       'Charter School',
       'Magnet School',
       'Grade 9 offered',
       'Grade 10 offered',
       'Grade 11 offered',
       'Grade 12 offered',
       'School ID',
       'Agency ID']
data.columns = cols
```

```python
# drop the columns that we do not need
data.drop(columns = ['School Type', 'Agency ID'], inplace = True)
```

```python
#replace each value of the cross, -, and double cross with '0'
#data.loc[0]['Magnet School']
x = data.loc[2]['Pupil/Teacher Ratio']
data.replace([x, '†', '‡'], '0', inplace=True)
```

```python
##convert columns to float so that I can do arithmetic with them
cols_to_convert = ['American Indian/Alaska Native',
       'Asian or Asian/Pacific Islander', 'Hispanic',
       'Black', 'White',
       'Nat. Hawaiian or Other Pacific Isl.', 'Two or More Races',
       'Free and Reduced Lunch', 'Total Students All Grades', 'Grades 9-12 Students',
       'Pupil/Teacher Ratio']  # Replace col1, col2, col3 with your actual column names

data[cols_to_convert] = data[cols_to_convert].astype(float)
```

```python
## The columns are counts right now for each race. Convert them to a percentage
races = ['American Indian/Alaska Native',
       'Asian or Asian/Pacific Islander', 'Hispanic',
       'Black', 'White',
       'Nat. Hawaiian or Other Pacific Isl.', 'Two or More Races']

for race in races:
    data[race] = round(data[race] / data['Total Students All Grades'], 3)
```

```python
# also convert the free lunch count to a free lunch proportion
data['Free and Reduced Lunch'] = round(data['Free and Reduced Lunch'] / data['Total Students All Grades'], 3)
```

```python
##convert the '1-Yes' '2-No' to simple binary 0 and 1

data = data.replace('1-Yes', 1)
data = data.replace('2-No', 0)
```

```python
# drop all of the rows where the schools don't have an address
data = data[~data['Address'].isnull()]
```

```python
cols_to_convert = ['Grade 9 offered', 'Grade 10 offered', 'Grade 11 offered',
       'Grade 12 offered', 'Charter School', 'Magnet School']
data[cols_to_convert] = data[cols_to_convert].astype(bool)
```

```python
data[data.isnull().any(axis=1)]
```

```python
## Right now if Total Students are reported as 0, then male students, female students,
##free lunches, and pupil/teacher ratio are reported as 0. change this to NaN
import numpy as np

columns_to_replace = ['Pupil/Teacher Ratio',
                     'Free and Reduced Lunch']  # List of columns to replace
data.loc[data['Total Students All Grades'] == 0, columns_to_replace] = np.nan
```

### Concat the 2 dfs

```python
df = pd.concat([data,private])
```

```python
df[df['High School'] == 1]
```

```python
#Format the City and Address
df['Address'] = df['Address'].str.title()
df['City'] = df['City'].str.title()
```

```python
#format the zipcode
df['Zipcode'] = df['Zipcode'].astype(int).astype(str)
```

```python
df = df[df['High School'] == 1]
```

```python
#change the type of charter school, magnet school, and high school
df[['Charter School', 'Magnet School']] = df[['Charter School', 'Magnet School']].astype(bool)
```

```python
#there should not be total students valued at 0, replace that with NaN
columns_to_replace = ['Total Students', 'Grades 9-12 Students']
df.loc[df['Total Students'] == 0, columns_to_replace] = np.nan
```

```python
#convert the racial proportions to racial percentages

races = ['American Indian/Alaska Native', 'Asian or Asian/Pacific Islander',
       'Hispanic', 'Black', 'White', 'Nat. Hawaiian or Other Pacific Isl.',
       'Two or More Races']

for race in races:
    df[race] = round(df[race] * 100, 3)
```
