```python
data = pd.read_csv('maxpreps_and_demo.csv')
```

```python
data.columns
```

```python
data.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
cols = ['Team 1', 'Team 1 Score', 'Team 2 Score', 'Outcome', 'Team 1 Boys Sports', 'T1 American Indian/Alaska Native',
       'T1 Asian or Asian/Pacific Islander', 'T1 Hispanic', 'T1 Black',
       'T1 White', 'T1 Nat. Hawaiian or Other Pacific Isl.',
       'T1 Two or More Races', 'T1 Total Students', 'T1 Grades 9-12 Students',
       'T1 Free and Reduced Lunch', 'T1 Pupil/Teacher Ratio',
       'T1 Charter School', 'T1 Magnet School', 'T1 Private School', 'T1 Expenditures/Pupil', 'Team 1 Label']
T1_subset = data[cols]
```

```python
## delete the rows that have null values for team 1 and team 2 score
T1_subset.dropna(subset = ['Team 1 Score', 'Team 2 Score'], inplace = True)
```

```python
#make a margin of victory column
T1_subset['MV'] = T1_subset['Team 1 Score'] - T1_subset['Team 2 Score']
```

```python
#avg margin of victory column
team1_avg_margin = T1_subset.groupby('Team 1 Label')['MV'].mean().reset_index()
```

```python
T1_subset['num Boys Sports'] = T1_subset['Team 1 Boys Sports'].apply(lambda x: len(x.split(',')))
```

```python
cols = ['Team 1', 'Team 1 Label', 'T1 American Indian/Alaska Native',
       'T1 Asian or Asian/Pacific Islander', 'T1 Hispanic', 'T1 Black',
       'T1 White', 'T1 Nat. Hawaiian or Other Pacific Isl.',
       'T1 Two or More Races', 'T1 Total Students', 'T1 Grades 9-12 Students',
       'T1 Free and Reduced Lunch', 'T1 Pupil/Teacher Ratio',
       'T1 Charter School', 'T1 Magnet School', 'T1 Private School', 'T1 Expenditures/Pupil']
teams_and_demo = data[cols]
teams_and_demo.drop_duplicates(subset = ['Team 1 Label'], inplace = True)
```

```python
avg_mv = list(team1_avg_margin['MV'])
```

```python
teams_and_demo['MV'] = avg_mv
```

```python
## now lets get the win percentage as well
teams_and_demo
```

```python
total_games = data.groupby('Team 1 Label').size().reset_index(name='Total Games')
```

```python
len(total_games)
```

```python
total_games = list(total_games['Total Games'])
```

```python
teams_and_demo['Total Games'] = total_games
```

```python
## make a column that counts the total wins per team
data['Win'] = (data['Outcome'] == 'W').astype(int)
total_wins = list(data.groupby('Team 1 Label')['Win'].sum().reset_index(name='Wins')['Wins'])
```

```python
teams_and_demo['Total Wins'] = total_wins
```

```python
teams_and_demo['Win Percentage'] = round(teams_and_demo['Total Wins'] / teams_and_demo['Total Games'], 3) * 100
```

```python
teams_and_demo.insert(2,'num boys Sports', T1_subset['num Boys Sports'])
```
