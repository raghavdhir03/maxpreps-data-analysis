```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import concurrent.futures
import time
import os
```

```python
data = pd.read_csv('ak.csv')
```

```python
data
```

```python
data['Team 1'] = data['Team 1'].str.replace(r'\s*\([^()]*\)', '', regex=True)
```

```python
def fixName(string):
    if string[0] == string[1] and string[0].isupper():
        string = string[1:]
    return string
```

```python
#pattern = r'^([A-Z])\1(?=\s)'

# Extract the correct part of the string using the pattern
data['Team 2'] = data['Team 2'].apply(fixName)
```

```python
null_values = data.isnull()

# Sum the null values for each column
null_counts = null_values.sum()
null_counts
```

```python
# Identify rows with missing 'Team 2' location
missing_locations = data[data['Team 2 Address'].isnull()]

# Iterate over each row with missing location
for index, row in missing_locations.iterrows():
    # Find matching row based on score
    matching_row = data[(data['Team 1'] == row['Team 2']) &
                              (data['Team 2'] == row['Team 1']) &
                              (data['Team 1 Score'] == row['Team 2 Score']) &
                              (data['Team 2 Score'] == row['Team 1 Score'])]
    # If a matching row is found, fill in the missing location
    if not matching_row.empty:
        data.at[index, 'Team 2 Address'] = matching_row.iloc[0]['Team 1 Address']

# Now merged_df should have 'Team 2' locations filled in where possible
```
