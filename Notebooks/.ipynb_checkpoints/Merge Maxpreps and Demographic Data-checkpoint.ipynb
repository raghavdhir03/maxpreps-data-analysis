```python
same_addy = same_addy.sort_values(by = ['Address', 'Zipcode'])
```

```python
def get_max_total_students(group):
    # Filter out NaN values in the 'Total Students' column
    group_no_nan = group.dropna(subset=['Total Students'])
    
    # Check if there are any non-NaN values
    if not group_no_nan.empty:
        # Return the row with the highest non-NaN value in the 'Total Students' column
        return group_no_nan.loc[group_no_nan['Total Students'].idxmax()]
    else:
        # If all values are NaN, return the first row of the group
        return group.iloc[0]

# Apply the function to each group
max_total_students_df = same_addy.groupby(['Address', 'Zipcode']).apply(get_max_total_students)
```

```python
merged_df = pd.merge(X, demo, left_on = 'Team 1 Zipcode', right_on = 'Zipcode', how = 'left')
```

```python
def fuzzy_match(df, id_column, team_address_column, address_column, threshold=60):
    """
    Perform fuzzy matching between 'team_address_column' and 'address_column' within groups identified by 'id_column'
    in the DataFrame 'df'. Rows with similarity scores above the specified 'threshold' are considered a match.
    Returns a DataFrame containing the matched rows.
    """
    matches = pd.DataFrame()
    grouped = df.groupby(id_column)
    no_match = []

    # Iterate over groups
    for _, group_df in grouped:
        # Initialize a variable to keep track of whether a match is found
        match_found = False

        # Iterate over rows in the group
        for index, row in group_df.iterrows():
            # Compare 'team_address_column' with 'address_column' using fuzzy string matching
            similarity_score = fuzz.partial_ratio(row[team_address_column], row[address_column])

            # If similarity score is above the threshold, consider it a match
            if similarity_score >= threshold:
                # Append the matched row to the filtered DataFrame
                matches = pd.concat([matches, pd.DataFrame([row])])
                match_found = True
                break  # Stop comparing once a match is found

        # If no match is found for any row in the group, append all rows to filtered DataFrame
        if not match_found:
            no_match.append(row[id_column])

    return matches, no_match

# Example usage:
# matched_rows = fuzzy_match(df=my_dataframe, id_column='ID', team_address_column='Team 1 Address', address_column='Address')
```

```python
cond = merged_df.groupby('Team 1 ID')['Team 1 ID'].transform('size') > 1
```

```python
check = merged_df[~cond]
```

```python
matched_rows = fuzzy_match(df=check, id_column='Team 1 ID', team_address_column='Team 1 Address', address_column='Address')
```

```python
intial_match = matched_rows[0]
```

```python
subset = merged_df[cond]
```

```python
second_match = fuzzy_match(df=subset, id_column='Team 1 ID', team_address_column='Team 1 Address', address_column='Address')
```

```python
##concat the 2 dfs, ignore index and order by ID
right = pd.concat([intial_match, second_match[0]]).sort_values(by = 'Team 1 ID').reset_index(drop = True)
```

```python
#left join the maxpreps table and this address match table using 'Team 1 ID'
team1_demo = pd.merge(maxpreps, right, on='Team 1 ID', how = 'left').drop(columns = ['Team 1 Address_y',
                                                                        'Team 1 Zipcode_y',
                                                                        'School Name',
                                                                        'Address',
                                                                        'City',
                                                                        'State',
                                                                        'Zipcode'
                                                                        ]).rename(columns = {'Team 1 Zipcode_x':
                                                                                            'Team 1 Zipcode',
                                                                                            'Team 1 Address_x':
                                                                                            'Team 1 Address'})
```
