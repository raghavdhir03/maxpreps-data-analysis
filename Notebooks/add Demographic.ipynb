```python
from thefuzz import fuzz, process
def fuzzy_match(df, id_column, team_address_column, address_column, threshold=60):
    """
    Perform fuzzy matching between 'team_address_column' and 'address_column' within groups identified by 'id_column'
    in the DataFrame 'df'. Rows with similarity scores above the specified 'threshold' are considered a match.
    Returns a DataFrame containing the matched rows.
    """
    matches = pd.DataFrame()
    grouped = df.groupby(id_column)
    no_match = []

    # Iterate over groups
    for _, group_df in grouped:
        # Initialize a variable to keep track of whether a match is found
        match_found = False

        # Iterate over rows in the group
        for index, row in group_df.iterrows():
            # Compare 'team_address_column' with 'address_column' using fuzzy string matching
            similarity_score = fuzz.partial_ratio(row[team_address_column], row[address_column])

            # If similarity score is above the threshold, consider it a match
            if similarity_score >= threshold:
                # Append the matched row to the filtered DataFrame
                matches = pd.concat([matches, pd.DataFrame([row])])
                match_found = True
                break  # Stop comparing once a match is found

        # If no match is found for any row in the group, append all rows to filtered DataFrame
        if not match_found:
            no_match.append(row[id_column])

    return matches, no_match

# Example usage:
# matched_rows = fuzzy_match(df=my_dataframe, id_column='ID', team_address_column='Team 1 Address', address_column='Address')
```

```python
matches, no_matches = fuzzy_match(merged_df, 'Team 1 URL', 'Team 1 Address', 'Address')
```

```python
#make function that will rename the columns to indicate what team the data is for
def add_team_to_column(df, cols, team_to_add):
    for col in cols:
        df.rename(columns = {col: team_to_add + ' ' + col}, inplace=True)


    return df
```
